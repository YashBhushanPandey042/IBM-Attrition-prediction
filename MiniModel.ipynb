```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/kaggle/input/ibm-hr-analytics-attrition-dataset/WA_Fn-UseC_-HR-Employee-Attrition.csv")
df.head(3)
```

```python
df["Department"].unique()
```

```python
en_data=df[["Gender","MaritalStatus","Department","Attrition","OverTime"]]
en_data.head(2)
```

```python
pd.get_dummies(en_data).info()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe=OneHotEncoder(drop="first")
ar=ohe.fit_transform(en_data).toarray()
```

```python
df_encoded=pd.DataFrame(ar,columns=["Gender_Male","MaritalStatus_Married","MaritalStatus_Single","Department_Research & Development","Department_Sales","Attrition_Yes","OverTime_Yes"])
df_encoded.head(5)
```

```python
df_new=df.drop(["Gender","MaritalStatus","Department","Attrition","OverTime"],axis=1)
```

```python
df=pd.concat([df_new.reset_index(drop=True),df_encoded.reset_index(drop=True)],axis=1)
df
```

```python
x=df[["Age","Gender_Male","MaritalStatus_Married","MaritalStatus_Single","Department_Research & Development","Department_Sales","DistanceFromHome","EnvironmentSatisfaction","JobSatisfaction","MonthlyIncome","OverTime_Yes","PercentSalaryHike","TotalWorkingYears","TrainingTimesLastYear","YearsAtCompany","YearsInCurrentRole","YearsSinceLastPromotion","YearsWithCurrManager","NumCompaniesWorked"]]
y=df["Attrition_Yes"]
x
```

```python
df["Attrition_Yes"].value_counts()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lor=LogisticRegression(class_weight="balanced")
lor.fit(x_train,y_train)
```

```python
lor.score(x_test,y_test)*100
```

```python
lor.predict([[41,0.0,0.0,1.0,0.0,1.0,1,2,4,100000,1.0,11,8,0,6,4,0,5,8]])
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cf=confusion_matrix(y_test,lor.predict(x_test))
cf
```
